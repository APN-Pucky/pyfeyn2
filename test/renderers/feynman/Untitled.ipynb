```python
from pyfeyn2.feynmandiagram import FeynmanDiagram, Leg, Propagator, Vertex
from pyfeyn2.render.feynman import FeynmanRender
from pyfeyn2.render.dot import feynman_adjust_points
```

```python

v1 = Vertex(x=0.3,y=0.4)
v2 = Vertex(x=0.5,y=0.4)

fd = FeynmanDiagram().add(
    v1,v2,
     Propagator(type="photon",label=r"$\gamma$").connect(v1, v2),
    Leg(type="fermion",label=r"$\chi$").set_target(v1).set_xy(0.2, 0.5).set_incoming(),
    Leg(type="fermion").set_target(v1).set_xy(0.2, 0.3).set_incoming(),
    Leg(type="fermion").set_target(v2).set_xy(0.6, 0.5).set_outgoing(),
    Leg(type="fermion").set_target(v2).set_xy(0.6, 0.3).set_outgoing()
)
```

```python
FeynmanRender(fd).render(show=True)
```
